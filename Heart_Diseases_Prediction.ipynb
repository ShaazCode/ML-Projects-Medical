**Splitting the features and target column**

```python
# So as we know in our dataset Target is the output column so we will sepearte it.
```

```python
X = heart_diseases.drop(columns='target', axis=1)

Y = heart_diseases['target']
```

```python
print(X.to_string())
```

```python
# above we had deleted the target column successfully
```

```python
print(Y)
# here we can see there's a column of target which we're storing inside the Y
```

**Training & Testing the data **

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
# Test_size = 0.2 => 20% for testing and 80% for training
# Stratify = this help us to prevent with imbalanced data
# Random_state = you get the same split every split at samne time if you wont say it then every split will differently
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
# so here we're gettting to know that x is having total 303 values then x train is containing 242 values and x test is containing 61 values
```

**Model Training**

```python
from sklearn.linear_model import LogisticRegression
```

```python
# logistic Regression model

model = LogisticRegression()
```

```python
# Training the Logistic regression model

# we can make use of train.fit to train the logisistic model
```

```python
model.fit(X_train, Y_train)
```

```python
# above we had trained the model quickly becuase the dataset is very low
```

**Model Evaluation**

```python
from sklearn.metrics import accuracy_score
```

```python
# Accuracy of training data
```

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print("Accuracy on training data :", training_data_accuracy)
```

So here we're getting the accuracy as 0.8512 it means the accuracy is 85%
somehow the 85% accuracy is also good but we have to boost it

```python
# Accuracy for test data

X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("Accuracy of test data:", test_data_accuracy)
```

So here the accuracy on test data is 81%

```python
# Training accuracy means => How the model trained from the given data

# Testing accuracy means => How well the model perform on new data unseen data
```

Building a predictive model

```python
input_data = (61,1,2,150,243,1,1,137,1,1.0,1,0,2)

# here the input data is into tuple convert it into array

input_data_as_numpy_array = np.asarray(input_data)


# reshape the numpy array as we are predicting

input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)

prediction = model.predict(input_data_reshaped)

print(prediction)
```

```python
if(prediction[0] == 0):
  print("The person does not have heart disease")
else:
  print("The person have heart diseases")
```
